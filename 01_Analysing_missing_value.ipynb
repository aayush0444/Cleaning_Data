```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data=pd.read_csv(r"df1_loan.csv")
data.head(5)
```

```python
data.shape
```

```python
data.isnull()
```

## **Finding the total null values in a data set using isnull and sum function**

```python
(data.isnull().sum()/data.shape[0])*100
```

## **Total null values in percent**

```python
(data.isnull().sum().sum()/(data.shape[0]*data.shape[1]))*100
```
