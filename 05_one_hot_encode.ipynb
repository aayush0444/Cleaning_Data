```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# Select categorical columns
endf = df[['Gender', 'Married']].copy()
endf = endf.fillna(endf.mode().iloc[0])  # Fill missing values with mode

# Initialize encoder
ohe = OneHotEncoder(drop='first')

# Fit and transform
encoded_array = ohe.fit_transform(endf)

# âœ… Get correct column names from encoder
columns = ohe.get_feature_names_out(['Gender', 'Married'])

# Create DataFrame
encoded_df = pd.DataFrame(encoded_array, columns=columns)

print(encoded_df.head())
```
